```python
df5 = pd.read_csv(r"C:\Users\shiva\Downloads\Cars (4).csv")
df5
```

```python
MPG = df5['MPG']
prob_a = (MPG > 38).mean()
prob_b = (MPG < 40).mean()
prob_c = ((20 < MPG) & (MPG < 50)).mean()

print(f'P(MPG > 38):{prob_a:.4f}')
print(f'P(MPG < 40):{prob_b:.4f}')
print(f'P(20 < MPG < 50):{prob_c:.4f}')
```

# q 24

```python
import scipy.stats as stats
import math

sample_mean = 260
population_mean = 270
sample_std_dev = 90
sample_size = 18

t_score = (sample_mean - population_mean) / (sample_std_dev / math.sqrt(sample_size))
probability = stats.t.cdf(t_score, df=sample_size - 1)

print(f"T-Score: {t_score}")
print(f"Probability: {probability:.4f}")
```
